## üìà Load-Depth Curve: Sample Indent 49

This Excel file contains nanoindentation data:

- `Depth`: indentation depth in nanometers
- `Load`: applied load in micro-Newtons

The curve provides mechanical response information for a single indent and can be used to compute hardness, modulus, or detect anomalies.

This example shows how to load `.xlsx` using `pandas` and visualize it using `matplotlib`.


üîπ 1. üì• Download via gdown

```python
# Download Excel file using gdown
import gdown

gdown.download(id="1WBUc8i6tZcy519C4DYsusIIMBEInHgUY", output="49.xlsx", quiet=False)
```

üîπ 2. üìä Load and Plot Using Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_excel("49.xlsx")

# Preview
display(df.head())

depth = pd.to_numeric(df.iloc[:, 0], errors='coerce')  # First column
load = pd.to_numeric(df.iloc[:, 1], errors='coerce')   # Second column

plt.figure(figsize=(6, 4))
plt.plot(depth, load, marker='o')
plt.xlabel('Depth (nm)')
plt.ylabel('Load (\u00B5N)')
plt.title('Load-Depth Curve: Indent 49')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## üì¶ Reading VK4 Files: Full Image Extraction

The `.vk4` format stores data from Keyence VK laser microscopes, including:

- `height`: 2D surface profile (depth map)
- `light`: raw laser intensity image
- `rgb`: optical RGB image
- `rgb+light`: overlay of laser intensity and RGB

This section demonstrates how to:
1. Read the `.vk4` file using `VkContainer`
2. Extract **all image types**
3. Plot them directly without cropping

üîß Code Cell: Read and Display All Image Types

```python
import VkContainer
import numpy as np
import matplotlib.pyplot as plt

filename = "49.vk4"

# Load with full builder
with open(filename, 'rb') as f:
    builder = VkContainer.Vk4Builder(f)
    director = VkContainer.VkDirector(builder)
    vk4 = director.build()

# Get image dimensions
width, height = vk4.image_width, vk4.image_height
print(f"üìè Image dimensions: {width} √ó {height}")

# Define expected layers
selected_layers = {
    "Height Map": vk4.height_data,
    "Laser Intensity": vk4.light_intensity_data,
    "Laser + Optical Overlay": vk4.rgb_light_data
}

# Display images
for title, data in selected_layers.items():
    if data and "data" in data:
        raw = data["data"]
        size = raw.size

        # Determine shape
        try:
            if title == "Laser + Optical Overlay":
                img = np.reshape(raw, (height, width, 3))
            else:
                img = np.reshape(raw, (height, width))
        except Exception as e:
            print(f"‚ùå Could not reshape {title}: {e}")
            continue

        print(f"‚úÖ {title} ‚Äî shape: {img.shape}")
        plt.figure(figsize=(6, 5))
        plt.imshow(img, cmap='viridis' if img.ndim == 2 else None)
        if img.ndim == 2:
            plt.colorbar()
        plt.title(title)
        plt.axis('off')
        plt.tight_layout()
        plt.show()
    else:
        print(f"‚ö†Ô∏è {title} not found in this file.")
```
